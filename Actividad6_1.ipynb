# **Casas y Departamentos Entero**

## Caso 1

```python
sns.scatterplot(x='host_response_rate', y='host_acceptance_rate', color="blue", data=filtro1)
```

```python
from sklearn.linear_model import LinearRegression
Vars_Indep= filtro1[['host_response_rate']]
Var_Dep= filtro1['host_acceptance_rate']
model= LinearRegression()
type(model)
model.fit(X=Vars_Indep, y=Var_Dep)
det1= model.score(Vars_Indep,Var_Dep)
cor1 = np.sqrt(model.score(Vars_Indep,Var_Dep))
print('Coeficiente de Determinación: ',det1)
print('Coeficiente de Correlación: ',cor1)
```

Correlación positiva débil. Por lo tanto, si el host_response_rate sube, el host_acceptance_rate no necesariamente sube también en la base de datos de airbnb de Ciudad de México en tipos de cuarto de casas/departamentos enteros.
